4. The Instacart CFO isn’t interested in customers who don’t generate much revenue for the app. Create an exclusion flag for low-activity customers (customers with less than 5 orders) and exclude them from the data. Make sure you export this sample.

```python
# Count the number of orders per customer
order_counts = df_final_updated.groupby('user_id').size().reset_index(name='order_count')
```

```python
# Merge the order counts back into the original DataFrame
df_final_updated = df_final_updated.merge(order_counts, on='user_id')
```

```python
# Create an exclusion flag for low-activity customers (less than 5 orders)
df_final_updated['Exclusion_Flag'] = df_final_updated['order_count'] < 5
```

```python
# Filter out low-activity customers
df_final_updated_filtered = df_final_updated[df_final_updated['Exclusion_Flag'] == False]
```

```python
# Drop the Exclusion_Flag and order_count columns as they are no longer needed
df_final_updated_filtered = df_final_updated_filtered.drop(columns=['Exclusion_Flag', 'order_count'])
```

```python
# Export the filtered data to a pickle file
df_final_updated_filtered.to_pickle(os.path.join(drive_directory, '02 Data','Prepared Data', 'Final_data_updated.pkl'))
```
